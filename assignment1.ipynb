1.6 Implement np.sin(), log(), log2(), np.exp()

```python
import numpy as np

sample_array = np.array([1, 2, 3, 4, 5])

sin_result = np.sin(sample_array)
log_result = np.log(sample_array)
log2_result = np.log2(sample_array)
exp_result = np.exp(sample_array)

print("Sin:", sin_result)
print("Log:", log_result)
print("Log2:", log2_result)
print("Exponential:", exp_result)
```

```python
import matplotlib.pyplot as plt

people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40, 45]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

plt.figure(figsize=(8, 6))
plt.scatter(height, weight, s=100, c='blue', alpha=0.5, label='Health Data')

plt.title('Health Data Visualization')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

for i, person in enumerate(people):
    plt.annotate(person, (height[i], weight[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.legend()
plt.grid(True)
plt.show()
```

2.4 Bar Chart


```python
import matplotlib.pyplot as plt

people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40, 45]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

avg_age = sum(age) / len(age)
avg_height = sum(height) / len(height)
avg_weight = sum(weight) / len(weight)

categories = ['Age', 'Height', 'Weight']
averages = [avg_age, avg_height, avg_weight]

plt.figure(figsize=(8, 6))
plt.bar(categories, averages, color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Average Age, Height, and Weight')
plt.xlabel('Categories')
plt.ylabel('Average')
plt.grid(axis='y')
plt.show()
```

2.5 HISTOGRAM

```python
import matplotlib.pyplot as plt

people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40, 45]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

plt.figure(figsize=(8, 6))
plt.hist(age, bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(height, bins=5, color='lightgreen', edgecolor='black')
plt.title('Histogram of Height')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(weight, bins=5, color='salmon', edgecolor='black')
plt.title('Histogram of Weight')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
