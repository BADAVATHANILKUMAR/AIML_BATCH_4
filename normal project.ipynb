```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/emails.csv')
df.head()
```

```python
df = df.drop(df.columns[0], axis=1)
df.head()
```

```python
free_col = df.columns[100]
click_col = df.columns[200]
label_col = df.columns[-1]
```

```python
spam = df[df[label_col] == 1]
not_spam = df[df[label_col] == 0]
df.head()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(not_spam[free_col], not_spam[click_col], color='blue', alpha=0.5, label='Not Spam')
plt.scatter(spam[free_col], spam[click_col], color='red', alpha=0.5, label='Spam')
```

```python
x = df['free']
y = df['click']
plt.scatter(df['free'], df['click'])
plt.title("Word Count Scatter: 'Free' vs. 'Click' in Emails")
plt.xlabel("Count of 'free'")
plt.ylabel("Count of 'click'")
plt.legend()
plt.grid(True)
plt.show()
```
